```python
import requests 
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

```python
wiki_url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
table_class='wikitable sortable'
```

```python
response=requests.get(wiki_url)
soup=BeautifulSoup(response.text, 'html.parser')
```

```python
# Scrapping the Wiki page

table = soup.find('table',{'class':'wikitable sortable'})
table_rows = table.find_all('tr')

data = []
for row in table_rows:
    data.append([t.text.strip() for t in row.find_all('td')])

df = pd.DataFrame(data, columns=['PostalCode', 'Borough', 'Neighbourhood'])

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
df
```

```python
df = pd.DataFrame(data, columns=['PostalCode', 'Borough', 'Neighbourhood'])

# Ignore cells with a borough that is "Not assigned"

indexNames = df[(df['Borough'] == 'Not assigned')].index
df.drop(indexNames, inplace=True)  

df=df.drop(df.index[0])

df
```

```python
# Combining repeated postal codes into one row with coma separated neighbourhoods

df2 = df.groupby(['PostalCode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

df2
```

```python
# Importing the csv file conatining the latitudes and longitudes for various neighbourhoods in Canada

location_url = pd.read_csv('http://cocl.us/Geospatial_data')

location_url.head()
```

```python
# Merging the two tables for getting the Latitudes and Longitudes for various neighbourhoods in Canada

location_url.rename(columns={'Postal Code':'PostalCode'},inplace=True)
df3 = pd.merge(df2, location_url,on='PostalCode')
df3.head()
```

```python
# Getting all the rows from the data frame which contains Toronto in their Borough

df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

```python
# Visualizing all the Neighbourhoods of the above data frame using Folium

!pip install folium
import folium 
```

```python
# Visualizing all the Neighbourhoods of the above data frame using Folium

map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for Latitude,Longitude,Borough,Neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(Neighbourhood, Borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [Latitude,Longitude],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
# KMeans Clustering for Neighborhood

from sklearn.cluster import KMeans

k=5
toronto_clustering = df4.drop(['PostalCode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_

df4.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
# create map

import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors

map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for Latitude, Longitude, Neighbourhood, Cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(Cluster), parse_html=True)
    folium.CircleMarker(
        [Latitude, Longitude],
        radius=5,
        popup=label,
        color=rainbow[Cluster-1],
        fill=True,
        fill_color=rainbow[Cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
